### Estatística

- O que é?
    - Estatística é uma ciência que estuda a coleta, organização, análise e registro de dados por 
    amostras. Desde a Antiguidade, quando se registravam nascimentos e mortes, a estatística tem sido fundamental para tomar decisões. Ela fundamenta conclusões com base em estudos realizados.

- Para que serve?
    - A estatística é uma ciência que utiliza teorias probabilísticas para explicar a frequência da ocorrência de eventos(Ex.: Possibilidade de determinado time ganhar um campeonato). Ela é aplicada tanto em estudos observacionais quanto em experimentos, permitindo modelar a aleatoriedade e a incerteza. Seu objetivo é estimar ou possibilitar a previsão de acontecimentos futuros. 
   
    - A estatística serve para:
        - Coletar dados: Ela nos ajuda a obter informações precisas a partir de dados coletados.
        - Analisar informações: Com base nos dados, podemos entender fenômenos complexos.
        - Tomar decisões embasadas: Desde questões simples até complexas, a estatística orienta escolhas informadas.
    - Por exemplo, os dados sobre casos de COVID-19 são usados para tomar decisões em saúde pública. A estatística está presente em nosso cotidiano, guiando desde políticas até pesquisas por vacinas.

- Como fazer?
    - Fazer uma estatística envolve muitas etapas importantes.
        - Levante Hipóteses:
        - Comece definindo a hipótese que deseja confirmar ou investigar. 
    - Obtenção dos Dados:
        - Colete dados relevantes por meio de amostragem ou experimentos. 
        - Esses dados serão a base para sua análise.
    - Organização dos Dados:
        - Utilize planilhas ou softwares para organizar os dados coletados. 
        - Isso vai facilitar a análise posterior.
    - Análise dos Dados:
        - Calcule parâmetros estatísticos, como média, desvio padrão e outros, para entender a variação nos dados.
    - Interpretação dos Dados:
        - Crie gráficos e tabelas para visualizar os resultados. 
        - Discuta o significado dos dados e suas implicações.
    - Conclusão e Inferência:
        - Com base nos resultados, tire conclusões ou faça previsões relevantes.

### Média

- O que é?
    - A média é uma medida de tendência central que busca representar um valor típico ou representativo do conjunto de dados. Ela é calculada somando todos os valores de um conjunto e dividindo pelo número de elementos. 

- Para que serve?
    - A média nos dá uma ideia geral das características desse conjunto de números. 

- Como fazer?
    - Exemplo, se tivermos um grupo de idades (22, 20, 21, 24 e 20 anos), a média de idade desse grupo seria calculada como:
        - media = (22 + 20 + 21 + 24 + 28) / 5
        - media = (10 + 10 + 10) / 3


### Mediana

- O que é?
A mediana é como o “meio termo” de um conjunto de dados. Imagine que você tem uma lista de números organizados em ordem crescente ou decrescente. A mediana é o valor que fica exatamente no meio dessa lista. Ela divide o conjunto em duas partes iguais: metade dos números são menores que a mediana, e a outra metade é maior.

- Para que serve?
    - A mediana é uma medida estatística que serve para encontrar o valor central de um conjunto de dados. Ela é especialmente útil quando temos valores extremos ou quando a distribuição dos dados não é simétrica.
    - Algumas situações em que a mediana é aplicada:

        - Dados Assimétricos: Quando os dados não têm uma distribuição simétrica (como uma curva normal), a mediana é mais robusta do que a média. Ela não é afetada por valores extremos.
        - Renda e Salários: A mediana é frequentemente usada para representar a renda ou os salários de uma população. Ela mostra o valor que divide a população em duas partes iguais: metade ganha mais e metade ganha menos.
        - Ordenação de Dados: A mediana é útil para classificar dados em ordem crescente ou decrescente. Ela ajuda a identificar o valor central.
        - Dados Discretos: Quando trabalhamos com dados discretos (como contagens ou categorias), a mediana é preferível à média.
        - Em resumo, a mediana nos ajuda a entender a tendência central dos dados, considerando sua posição no conjunto.

- Como fazer?
    - A mediana é o valor central em um conjunto de dados ordenado. Para calcular a mediana, siga estes passos:

        - Organize os dados: Coloque os valores em ordem crescente ou decrescente.
        - Verifique a quantidade de dados:
        - Se for ímpar, a mediana é o valor do meio.
        - Se for par, a mediana é a média aritmética dos valores centrais.
- Exemplos:

    - Quantidade ímpar de dados:
        - Dado o conjunto A = {12, 4, 7, 23, 38}, organizamos em ordem: {4, 7, 12, 23, 38}.
        - A mediana é o valor do meio, que é 12.
    - Quantidade par de dados:
        - Alturas dos jogadores de vôlei: {1,83m; 1,87m; 1,97m; 1,99m; 2,01m; 2,05m}.
        - A mediana é a média aritmética dos valores centrais (1,97m e 1,99m), resultando em 1,98m.
    
### Obs.: 

- Organizar os dados em ordem crescente ou decrescente antes de calcular a mediana é importante para garantir que encontremos o valor central corretamente. Vamos entender o motivo:

    - Quantidade Ímpar de Dados:
        - Quando temos uma quantidade ímpar de dados, a mediana é o valor exatamente no meio da lista ordenada.
        - Organizar os dados nos permite identificar facilmente o valor central, pois ele estará na posição exata.
        - Por exemplo, se tivermos 5 valores, o terceiro valor será a mediana.
    - Quantidade Par de Dados:
        - Se não organizarmos os dados, calcular a mediana em um conjunto par pode ser confuso.
        - Organizando-os, encontramos os dois valores centrais e calculamos a média entre eles.
Por exemplo, se tivermos 6 valores, os valores do meio serão o terceiro e o quarto, e a mediana será a média entre eles.


### Moda

- O que é?
    - A moda é o valor mais frequente em um conjunto de dados. Em outras palavras, é o número que ocorre com maior frequência.

- Para que serve?
    - Ela nos ajuda a entender quais valores ocorrem com maior frequência em um conjunto de dados.
    - Por exemplo, se tivermos um conjunto de notas de alunos (5, 6, 7, 7, 8, 8, 8, 9), a moda seria 8, pois é o valor que aparece com mais frequência. 
    - A moda é especialmente útil quando queremos destacar tendências ou padrões em dados observados.

- Como fazer?
    - Anote os números do conjunto de dados.
    - Registre todos os valores disponíveis. Se houver muitos dados, use uma planilha no Excel ou papel e lápis para facilitar.
    - Ordene os números do menor ao maior.
    - Organize os valores em ordem crescente ou decrescente. Isso ajuda a identificar os números iguais em sequência direta.
    - Conte quantas vezes cada número se repete.
    - Determine quantas vezes cada valor aparece no conjunto de dados.
    - Identifique os valores mais frequentes.
    - Os números que aparecem com maior frequência é são a moda.
    - Lembre-se de que a moda é o valor mais comum em um conjunto de dados.
- Exemplo: 
    - Em uma sapataria, foram vendidos os seguintes números de sapato em um dia: 
    34, 39, 36, 35, 37, 40, 36, 38, 36, 38 e 41
    - O número 36 foi o mais frequente (3 pares), então a moda é igual a 36.

### Quartis

- O que é?
    - Os quartis são valores que dividem um conjunto de dados ordenados em quatro partes iguais. Eles são usados para entender a dispersão e a distribuição dos dados. Cada quartil representa uma porcentagem de dados:

    - Primeiro Quartil (Q1): Divide os 25% inferiores dos dados.
    - Segundo Quartil (Q2): É a mediana, dividindo os 50% medianos.
    - Terceiro Quartil (Q3): Divide os 25% superiores dos dados

- Para que serve?
    - Quartis desempenham um papel fundamental na análise estatística, permitindo uma avaliação detalhada da distribuição dos dados. 
    
    - Eles são úteis para:

    - Identificar Valores Extremos: Os quartis ajudam a detectar valores atípicos ou discrepantes nos dados.
    - Avaliar Simetria e Dispersão: Permitem entender a variabilidade e a forma da distribuição.
    - Comparar Grupos ou Amostras: Ao analisar quartis, podemos comparar diferentes conjuntos de dados.
    - Tomar Decisões Estratégicas: Na área de negócios, os quartis embasam decisões sobre metas de vendas, desempenho de equipes e oportunidades de mercado.

- Como fazer?

    - Para calcular os quartis de um conjunto de dados, prossiga com estes passos:

    - Organize os dados em ordem crescente ou decrescente.
    - Encontre a mediana (Q2), que é o segundo quartil.
    - Divida o conjunto de dados ordenado em duas metades:
    - O primeiro quartil (Q1) é a mediana da metade inferior dos dados.
    - O terceiro quartil (Q3) é a mediana da metade superior dos dados.
- Aqui estão as fórmulas gerais para calcular os quartis:

    - Primeiro Quartil (Q1):
        - Se o conjunto de dados tiver quantidade ímpar, Q1 é o valor exatamente no meio da metade inferior.
        - Se for quantidade par, Q1 é a média entre os dois valores centrais.
    - Terceiro Quartil (Q3):
        - Se o conjunto de dados tiver quantidade ímpar, Q3 é o valor exatamente no meio da metade superior.
        - Se for quantidade par, Q3 é a média entre os dois valores centrais.
- Exemplo: 
- Suponha que temos as notas dos alunos em uma prova de Estatística:

    - Notas: 5, 6, 7, 7, 8, 8, 8, 9
    - Organize os valores em ordem crescente:
    - 5, 6, 7, 7, 8, 8, 8, 9
    - Calcule a mediana (2º quartil):
    - Como temos 8 valores, a mediana é a média dos valores centrais: (7 + 8) / 2 = 7,5.
    - Primeiro Quartil (Q1):
    - É o valor entre o primeiro dado (5) e a mediana (7,5). Não temos exatamente um valor no meio, então fazemos a média entre eles: (5 + 7,5) / 2 = 6,25.
    - Terceiro Quartil (Q3):
    - É o valor entre a mediana (7,5) e o último dado (9). 
    - Novamente, fazemos a média: (7,5 + 9) / 2 = 8,25.
- Portanto, os quartis são:

- Q1: 6,25
- Q2 (mediana): 7,5
- Q3: 8,25

### Desvio padrão

- O que é?
    - O desvio-padrão é uma medida estatística que quantifica a variação ou dispersão de um conjunto de dados. Ele indica o grau de variação dos valores em relação à média aritmética desse conjunto. Em outras palavras, o desvio-padrão nos ajuda a entender o quão “espalhados” ou “concentrados” estão os dados em um conjunto.

- Para que serve?
    - O desvio-padrão é uma medida estatística que quantifica a variação ou dispersão de um conjunto de dados em relação à sua média aritmética. Ele nos ajuda a entender o quão “espalhados” ou “concentrados” estão os dados. Quanto maior o valor do desvio-padrão, maior a variabilidade dos dados em relação à média, e vice-versa. Em resumo, o desvio-padrão é uma ferramenta essencial para avaliar a consistência e a previsibilidade dos dados.

- Como fazer?
    - O desvio-padrão é uma medida que expressa o grau de dispersão de um conjunto de dados. Ele indica o quanto os valores se afastam da média aritmética desse conjunto. Quanto mais próximo de zero for o desvio-padrão, mais homogêneos são os dados. Para calcular o desvio-padrão, siga estes passos:

        - Calcule a média dos dados.
        - Subtraia a média de cada valor para obter as diferenças.
        - Eleve cada diferença ao quadrado.
        - Calcule a média dos quadrados das diferenças (variância).
        - Tire a raiz quadrada da variância para obter o desvio-padrão.
- Exemplo: se tivermos as alturas de uma equipe de remo (1,55 m; 1,70 m; 1,80 m), o cálculo seria:

    - Média (M<sub>A</sub>): (1,55 + 1,70 + 1,80) / 3 = 1,68 m
    - Diferenças: 1,55 - 1,68 = -0,13; 1,70 - 1,68 = 0,02; 1,80 - 1,68 = 0,12
    - Quadrados das diferenças: (-0,13)<sup>2</sup> + 0,02<sup>2</sup> + 0,12<sup>2</sup> = 0,0197
    - Variância (V): 0,0197 / 3 = 0,00657
    - Desvio-padrão (σ): √0,00657 ≈ 0,081 m
